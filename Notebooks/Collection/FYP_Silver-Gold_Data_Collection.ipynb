```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime

# Define the base URL and the date range
base_url = "https://www.bullion-rates.com/gold/PKR/{year}-{month:02d}-history.htm"
start_date = datetime(2007, 12, 1)
end_date = datetime(2025, 1, 24)

# Initialize an empty list to store the data
data = []

# Iterate through each month in the date range
current_date = start_date
while current_date <= end_date:
    year = current_date.year
    month = current_date.month
    url = base_url.format(year=year, month=month)
    print(f"Fetching data from: {url}")

    # Fetch and parse the page
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Failed to fetch data for {year}-{month:02d}")
        current_date = datetime(year + (month // 12), (month % 12) + 1, 1)
        continue

    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table rows containing the data
    rows = soup.find_all('tr', class_='DataRow')
    for row in rows:
        cols = row.find_all('td', class_='rate')
        if len(cols) == 3:
            date = cols[0].text.strip()
            silver_price_oz = cols[1].text.strip()
            silver_price_gram = cols[2].text.strip()
            data.append([date, silver_price_oz, silver_price_gram])

    # Move to the next month
    current_date = datetime(year + (month // 12), (month % 12) + 1, 1)

# Save the data to a CSV file
columns = ['Date', 'Gold Price (oz)', 'Gold Price (gram)']
df = pd.DataFrame(data, columns=columns)
df.to_csv('gold_prices.csv', index=False)
print("Data saved to gold_prices.csv")
```
