```python
!wget -P /kaggle/working -nc "https://raw.githubusercontent.com/MuhammadAreebKazmi2/FYP-FinSage-Data-Collection/refs/heads/main/Data/gold_prices.csv"
```

```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
df = pd.read_csv('/kaggle/working/gold_prices.csv')
df
```

```python
# Load the data
file_path = "/kaggle/working/gold_prices.csv"
data = pd.read_csv(file_path)

# Clean the data
# Remove commas and convert prices to numeric format
data["Gold Price (oz)"] = data["Gold Price (oz)"].str.replace(",", "").astype(float)
data["Gold Price (gram)"] = data["Gold Price (gram)"].str.replace(",", "").astype(float)

data = data[data["Date"] >= "2013-01-01"]

# Convert the Date column to datetime format
data["Date"] = pd.to_datetime(data["Date"])

# Set the Date column as the index
data.set_index("Date", inplace=True)


# Resample the data to get the end-of-month values
monthly_data = data.resample('M').last()

# Calculate monthly returns
# monthly_data['Monthly Return (oz)'] = monthly_data['Gold Price (oz)'].pct_change()
monthly_data['Monthly Return'] = monthly_data['Gold Price (gram)'].pct_change()

# Save the data to a new CSV file
output_file_path = "gold_prices_monthly.csv"
monthly_data.to_csv(output_file_path)

print(f"Monthly data with returns saved to {output_file_path}")
monthly_data
```

# KSE 100 Index Monthly

```python
# Load the data
file_path = "/kaggle/working/Karachi 100 Historical Data - Daily.csv"
data = pd.read_csv(file_path)
data
```

```python
import pandas as pd

# Load the data
file_path = "/kaggle/working/Karachi 100 Historical Data - Daily.csv"
data = pd.read_csv(file_path)

# Clean the data
# Remove commas and convert numeric columns to float
data["Price"] = data["Price"].str.replace(",", "").astype(float)
data["Open"] = data["Open"].str.replace(",", "").astype(float)
data["High"] = data["High"].str.replace(",", "").astype(float)
data["Low"] = data["Low"].str.replace(",", "").astype(float)
data = data[data["Date"] >= "2013-01-01"]
# Convert the Date column to datetime format
data["Date"] = pd.to_datetime(data["Date"])

# Set the Date column as the index
data.set_index("Date", inplace=True)

# Resample the data to get the end-of-month values
monthly_data = data.resample('M').last()

# Calculate monthly returns
monthly_data['Monthly Return'] = monthly_data['Price'].pct_change()

# Save the data to a new CSV file
output_file_path = "kse100_monthly_data.csv"
monthly_data.to_csv(output_file_path)

print(f"Monthly data with returns saved to {output_file_path}")
```
